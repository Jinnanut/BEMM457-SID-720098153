```python
import pandas as pd

# Load the data to examine its structure
file_path = '/Users/jinnanut/Library/Mobile Documents/com~apple~CloudDocs/NUT/Exeter University/Master/BEMM457 Topic in BA/Individual Project/Datacoding/DataQ1.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset to understand its structure
# data.head()

import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

# Rename columns for clarity
data.columns = ['Spending', 'Region', 'Purpose', 'Average_Nights_Stay']

# Define dependent and independent variables
X = data[['Region', 'Purpose', 'Average_Nights_Stay']]
y = data['Spending']

# Add a constant to the independent variables
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()

# Summary of the regression model
regression_summary = model.summary()

regression_summary
```

```python
# Regression plot for Spending and Region
sns.regplot(x=data['Region'], y=data['Spending'], line_kws={'color': 'red'})
plt.title('Regression Plot: Region vs Spending')
plt.xlabel('Region')
plt.ylabel('Spending per Day')
plt.show()

# Regression plot for Spending and Purpose
sns.regplot(x=data['Purpose'], y=data['Spending'], line_kws={'color': 'red'})
plt.title('Regression Plot: Purpose vs Spending')
plt.xlabel('Purpose')
plt.ylabel('Spending per Day')
plt.show()

# Regression plot for Spending and Average Nights Stay
sns.regplot(x=data['Average_Nights_Stay'], y=data['Spending'], line_kws={'color': 'red'})
plt.title('Regression Plot: Average Nights Stay vs Spending')
plt.xlabel('Average Nights Stay')
plt.ylabel('Spending per Day')
plt.show()
```

```python
# Compute the correlation matrix
correlation_matrix = data.corr()

# Visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Correlation Matrix')
plt.show()

correlation_matrix
```
